```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.array([1,2,3,4])
```

```python
y = np.array([5,6,7,8])
```

```python
plt.plot(x,y,ls='--',color='r')
font2 = {'family':'serif','color':'darkred','size':15}
plt.subplot(1,2,2)
plt.title('My Record',fontdict=font2,loc='left')
plt.xlabel('Temperature')
plt.ylabel('Celsius')
plt.grid(axis='y')
```
